### CIFAR-10 Photo Classification Dataset

CIFAR-10 dataset is for image classification. It consists of 60,000 images of 10 classes. In total, there are 50,000 training images and 10,000 test images. The dataset is divided into 6 parts – 5 training batches and 1 test batch. Each batch has 10,000 images.
It is comprised of 60,000 32×32 pixel color photographs of objects from 10 classes.

## Load Dataset

The load_dataset() function implements these behaviors and can be used to load the dataset.

```python
def load_dataset():
    #load train and test dataset
    (trainX,trainY),(textX,textY) = cifar10.load_data()
    
    #Onehotencode the labels
    trainY = to_categorical(trainY)
    testY = to_categorical(trainY)
    
    return trainX,trainY,textX,textY
```

##### Prepare Pixel Data

The pixel values for each image in the dataset are unsigned integers in the range between no color and full color, or 0 and 255.

We do not know the best way to scale the pixel values for modeling, but we know that some scaling will be required.

A good starting point is to normalize the pixel values, e.g. rescale them to the range [0,1]. This involves first converting the data type from unsigned integers to floats, then dividing the pixel values by the maximum value.

```python
#Convert integers to float
trainX_norm=trainX.astype('float32')
testX_norm=testX.astype('float32')

#normalize range to 0 to 1
trainX_norm=trainX_norm/255.0
testX_norm=testX_norm/255.0

```
